```python
import pandas as pd
from sts.stssig.PVA.coil import bbcomp, trendvol, bb, trend, coil_ts, atr
from sts.mktdata.ticker import Ticker
from sts.quant.candle import Candle
pd.options.plotting.backend = 'plotly'
```

```python
ticker = Ticker('us/etf.yml')
```

```python
df = ticker.history('USO', '20240101', '20260409').set_index('ts')
```

```python
fig = Candle(df).plot()
fig.update_layout(height = 600)
```

```python
log_df = pd.DataFrame(Candle(df).log())
```

```python
import numpy as np
scale = 5
log_df['coil_bbcomp'] = bbcomp(log_df, window = 20 * 5, std = 2, coil_window = 2 * scale, coil_low_score = 0.3, coil_buffer = 0.3)
log_df['coil_trendvol'] = trendvol(log_df, short = 10 * scale, coil_window = 5 * scale, coil_low_tscore = 0.3, coil_low_vol = 0.3 , coil_buffer = 0.3)
```

```python
fig = log_df.plot(y = ['coil_bbcomp', 'coil_trendvol'])
fig.update_layout( height = 400)
```

```python
from sts.stssig.PVA.base import siglen
```

```python

log_df['coil_bbcomp_len'] = siglen(log_df['coil_bbcomp'], 0.2)
log_df['coil_trendvol_len'] = siglen(log_df['coil_trendvol'], 0.2)
```

```python
log_df.plot( y = ['coil_bbcomp_len', 'coil_trendvol_len'])
```
