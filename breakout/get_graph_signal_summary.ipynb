```python
from pathlib import Path
import pandas as pd
folder = Path('~/data/stsdata/research/usstock/graphsignal/').expanduser().absolute()
result = []
for f in folder.iterdir():
    df = pd.read_parquet(str(f))
    res = df.reset_index().iloc[-1].to_dict()
    res['symbol'] = f.name
    result.append(res)

if len(result) > 0:
    result = pd.DataFrame(result)

result.to_csv('~/Dropbox/stsinvest/research/stock_graph_signal.csv')
```
