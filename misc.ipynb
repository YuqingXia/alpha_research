```python
from sts.utils import configParser
from sts.data.ticker import Ticker
from sts.quant.candle import Candle
import pandas as pd
pd.options.plotting.backend = 'plotly'
```

### Graph signal output to dropbox

```python
asset = 'usstock'
ticker = Ticker('us/etf.yml' if asset == 'usetf' else 'us/stock.yml')
df = pd.read_parquet(f'~/data/stsdata/research/{asset}/graphsignal')
sym_industry = ticker.symbol_industry()
df = df.groupby('symbol').last().reset_index()
df = df.merge(sym_industry[['symbol', 'sector', 'name']].drop_duplicates('symbol'), on = 'symbol', how = 'left')
df.sort_values('trendvol250', ascending=False).to_csv(f'~/Dropbox/stsinvest/research/{asset}_graph_signal.csv')
```
