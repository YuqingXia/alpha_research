```python
# import packages
import yfinance as yf
import pandas as pd
import numpy as np
import datetime
```

```python
ticker = yf.Ticker("SPY")
```

```python
strike = ticker.history(period="1W")["Close"].iloc[0]
```

```python
option_chain = ticker.option_chain("2023-10-06")
```

```python
calls = option_chain.calls
puts = option_chain.puts
```

```python
puts.openInterest.sum() / calls.openInterest.sum()
```

```python
(puts[puts.strike <= strike]).openInterest.sum() / (calls[calls.strike >= strike]).openInterest.sum()
```
