# Backtest Data Prepare

```python
from sts import dio
import datetime as dt
import pandas as pd
```

```python
def pre_stsbt_stock_market_data(sym: str, start: dt.datetime = None, end: dt.datetime = None, fundingRate: float = 0):
    ticker = dio.equity.Ticker(sym)
    df = ticker.history(start=start, end=end)
    df = df[["Close"]].copy()
    df["SymKey"] = sym
    df["Beta"] = 1
    df["Date"] = df.index
    df["FundingRate"] = fundingRate
    df["FungingTimeTS"] = (df["Date"] - df["Date"].iloc[0]).dt.days / 365.0
    df["Multiplier"] = 1.0
    df["TradeCost"] = 0.0
    df["Fx"] = 1.0

    return df
```

```python
sym = "SPY"
start = dt.datetime(2024, 10, 1)
end = dt.datetime(2024, 10, 10)
market_df = pre_stsbt_stock_market_data(sym, start=start, end=end)
```
