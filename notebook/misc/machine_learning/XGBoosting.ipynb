```python
from xgboost import XGBRegressor
from sklearn.inspection import partial_dependence, PartialDependenceDisplay
import numpy as np
```

```python
def get_plot(X, y):
    model = XGBRegressor(
        objective="reg:squarederror",
        eta=0.3,
        gamma=0,
        max_depth=5,
        min_child_weight=1,
        subsample=0.8,
        colsample_bylevel=1,
        reg_alpha=0,
        reg_lambda=1,
    )

    model.fit(X, y)

    pd_result = partial_dependence(model, X, [0, 1], grid_resolution=10)

    display = PartialDependenceDisplay(
        [pd_result],
        features=[(0, 1)],
        feature_names=["X1", "X2"],
        target_idx=0,
        deciles={0: np.linspace(0, 1, 5), 1: np.linspace(0, 1, 5)},
    )

    display.plot()
```

y = X1 + X2

```python
# code
X = np.random.rand(100_000, 2)
epsilon = np.random.rand(X.shape[0])
y = X[:, 0] + X[:, 1] + epsilon * 0.5
get_plot(X, y)
```

y = X1 * X2

```python
# code
X = np.random.rand(100_000, 2)
epsilon = np.random.rand(X.shape[0])
y = X[:, 0] * X[:, 1] + epsilon * 0.5
get_plot(X, y)
```

y = X1 / (X2 + 0.5)

```python
# code
X = np.random.rand(100_000, 2)
epsilon = np.random.rand(X.shape[0])
y = X[:, 0] / (X[:, 1] + 0.5) + epsilon * 0.5
get_plot(X, y)
```

y = X1 * sign(X2 - 0.5)

```python
# code
X = np.random.rand(100_000, 2)
epsilon = np.random.rand(X.shape[0])
y = X[:, 0] * np.sign((X[:, 1] - 0.5)) + epsilon * 0.5
get_plot(X, y)
```

```python
# code
X = np.random.rand(100_000, 2)
epsilon = np.random.rand(X.shape[0])
y = X[:, 0] * np.sin(X[:, 1] * 10) + epsilon * 0.5
get_plot(X, y)
```
