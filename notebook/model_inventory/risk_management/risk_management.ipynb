## Risk Sizing

Risk Sizing, profit taking, and stoploss are actually try to answer the same question, given the estimated win/loss probability, how much risk shall we hold. Specifically

1. We have stoploss price, which actually means that we believe once price pass that level, the return probability is 0 and thus we should not holding our position.
2. Similarly, some investor has a predefined profit taking level because they believe the price cross that level is zero and better to unwinde the position at that level.
3. Another popular way is gradually cut your size after with moderest profit. The underlying assumption is that we believe either the risk reward ratio become low after some profit or the probability of win become smalller, thus we should cut our size. 

All these three scenarios can be answered in the same framework:
1. Given a tolerant risk profile
2. With estimated win probability and lose probability
what is the optimal size we should keep. This problem is already studied by Kelly who gave following fomula

$$
\text{Best Size as \% target risk} = \frac{(\text{reward to risk ratio})( \text{probability of winning} ) - (\text{probability of losing})}{\text{reward to risk ratio}}
$$

## Kelly Kritieria 

Assume each bet we always take a fraction $f$ from total wealth, the reward if win is $rf$ and lose is $f$ if loss. After total $N$ bet, assume there is $i$ times win and $j$ times lose with $i+j = N$, then the total wealth after $N$ bet is $(1+rf)^i(1-f)^j$. Assume the average growth rate is $\alpha$, then 
$$\begin{aligned}
(1+\alpha)^N & = (1+rf)^i(1-f)^j \\
\alpha & = (1+rf)^{i/N})(1-f)^{j/N} - 1\\
\alpha & = (1+rf)^p(1-f)^q -1 \\
\end{aligned}$$

In order to maximize $\alpha$, we can maximize (1+\alpha) and thus equivelent to maximize the log wealth growth which is 
$$
G(f)  = p\cdot log(1+rf) + q log(1-f)
$$

solve the maximization problem, we can get the optimal $\tilde f = \frac{rp - q}{r}$

From above process, one of the assumption is that on each bet, we always take a bet with fraction $f$ of total wealth, which is not always true in reality especially when your wealth become big. But this problem can be solved by first split your wealth into small group and within each group we can still apply Kelly Kriteria

Also in realty, it is hard to estimate the probability accurately and things change continously. This formula is more like to bet the result when probability does not change. If we can bet in infinitesimal time horizon, the Kelly Kriteria still apply, but in reality, it is not, we have constrain and we do not want to trade frequently due to the cost. Due to the convextity of the exponential function, the total return is smaller compared to bet at infinitesimal time interval. And thus we should not allocate such big risk with Kelly Kriteria, instead we can adjust it slightly. The longer the horizon we hold, the smaller the risk based on this formula. So we can cut to half, 1 3rd or 1 quarter based on the time horizon and risk volatility.

In my oppinion, the portfolio optimization should always do at strategy level, we cannot mix the position from different strategies. For example, a PM have a large beta exposure on an intraday strategy while he also have a large beta exposure at weekly rebalance strategy. Shall he be freaked out due to large risk exposure? But what I mean for risk allocation we should do at strategy first and then use Kelly Kriteria to bet for within each strategy.
