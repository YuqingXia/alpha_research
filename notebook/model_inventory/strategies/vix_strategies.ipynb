Vix carry strategy is a short vol strategy, which is negatively skewed. So for such strategy, the key part is to identify potential events that will disturbance the market. Such event including expected event like election, FOMC, geopolitical risk, and accumulated risk in stock after quiet bull regimes. 

The other strategy is long vol strategy, which is opposite to vix carry strategy. If we can have a signal with such big vol, then we find regimes works for both regimes. The most important thing for such strategy is to follow the market narrative especially macro story. 

In this note, we show the steps to improve the vix strategy.
