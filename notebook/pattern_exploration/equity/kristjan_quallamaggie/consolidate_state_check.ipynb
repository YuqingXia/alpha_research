## Consolidate Signal Selected kqtrends

```python
from sts.strategies.trending.kqtrend.consolidate import Consolidate
```

```python
ticker = Ticker("MMM")
df = Candle(ticker.history(period="10y")).log()
```

```python
consolidate = Consolidate(window=20, bb_buffer_threshold=0.2, mm_buffer_threshold=0.2)
consolidate_status = consolidate(df)
```

```python
pd.options.plotting.backend = "plotly"
for score_horizon, sample_horizon, idx in [
    ("DailyScore", "1D", -100),
    ("WeeklyScore", "1W", -100),
    ("MonthlyScore", "1ME", -100),
]:
    fig = consolidate_status[score_horizon].iloc[idx:].plot(width=1000)
    fig.show()
    fig = Candle(df.resample(sample_horizon).iloc[idx:]).plot(width=1000)
    fig.show()
```
