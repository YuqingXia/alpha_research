```python
from sts.dio.equity import Ticker
from sts.dio.symbols import sp500_symbol_table
from sts.strategy.trending import kqtrend
from sts.quant.candle import Candle
import pandas as pd
```

```python
ticker = Ticker("MMM")
df = ticker.history(period="3y")
df = Candle(df).log()
```

```python
breakout = kqtrend.Breakout(window=20, bb_buffer_threshold=0.2, mm_buffer_threshold=0.2)
breakout_status = breakout(df)
```

```python
pd.options.plotting.backend = "plotly"
breakout_status["WeeklyScore"].plot()
```
