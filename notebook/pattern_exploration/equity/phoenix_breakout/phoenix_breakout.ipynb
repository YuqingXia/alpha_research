# Setup

For any given time interval, we stick to the 20 moving average with 2 sigma bolling band. 

- [x] it should have a big decline : trending: at monthly or weekly level
- [x] it should have a consolidated regime: done, based on weekly and monthly consolidate indicators
- [x] up break trying: daily level, short term trending
- [ ] resistance level

```python
from sts.quant.indicators.trend import ma_trend
from sts.dio.equity import Ticker
from sts.dio.symbols import sp500_symbol_table
from sts.plots import plot_multi_time_series
from sts.quant.candle import Candle
import pandas as pd
import numpy as np
```

```python
trend_scores_scan = []
for sym in sp500_symbol_table.to_symbol():
    sym = sym.YahooSym
    print(sym, end=", ")
    try:
        ticker = Ticker(sym)
        df = ticker.history(period="8y")
        trend_scores = ma_trend.trend_score_multi_horizon(df, short=5, long=20)
        latest_score = {k: v.iloc[-1] for k, v in trend_scores.items()}
        latest_score["sym"] = sym
        trend_scores_scan.append(latest_score)
    except Exception as e:
        print(e)
```

```python
trend_scores_scan_df = pd.DataFrame(trend_scores_scan)
```

```python
trend_scores_scan_df[
    (trend_scores_scan_df["MonthlyScore"] < 0) & (trend_scores_scan_df["WeeklyScore"] > 0)
].sort_values("MonthlyScore")
```
