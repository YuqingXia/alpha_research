# Theory

Resistance in stock formed due to reasons

* psychology effects: stock market price action is formed by human behavior. Such resistance level is watched by most investors and thus is self strengthened.
* position effects: at resistance level, a lots of position formed at those level. The longer the resistance level touched , the stronger the level. Those trapped position and position build at lower level will tend to sell at the resistance level. But once break up, this will switch to a support level. 

So in this theory, the horizon resistance level is the most import level. While some people also use upward or downward straightline but it is mainly work as psychological effects. Position effect is not that strong. 

# Horizontal Resistance Level

we only focus on the base accumulation period. so it must first be a consolidation period and then we try to extract the maximum line. then we can extract the maximum and small value of each period to get rolling support and resistance level

```python
from sts.dio.equity import Ticker
from sts.quant.candle import Candle
from sts.quant.indicators.consolidate import bb_consoldiate
import pandas as pd
```

```python
ticker = Ticker("QQQ")
df = Candle(ticker.history(period="5y"))
```

```python
df.resample("1W").plot(width=800)
```

```python
weekly_df = df.resample("1W")
weekly_cons_score = bb_consoldiate.consolidation_score(weekly_df)
```

```python
pd.options.plotting.backend = "plotly"
weekly_cons_score.plot(width=800)
```

```python
weekly_cons_score = weekly_cons_score.to_frame()
```

```python
weekly_cons_score["Session"] = (weekly_cons_score["ConsScore"] < 1) * 1.0
```
