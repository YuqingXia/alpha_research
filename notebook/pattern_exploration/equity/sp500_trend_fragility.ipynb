```python
from sts.data.datatable.tables import sp500_datatable
from sts.quant.indicators.trend import trend_ewm
from sts.data.dapi import Ticker
import pandas as pd
```

we can first calculate the normalized price trend, then calculate average as well as the divergence. 
For the average, we can do 

1. simple average
2. capital average

```python
df_all = []
period = "10y"
short = 5
long = 20
for sym in sp500_datatable.keys():
    ticker = sp500_datatable[sym]
    df = ticker.history(period=period)
    if len(df) > 0:
        trend_strength = trend_ewm(df, short, long)
        trend_strength.name = str(sym)
        df_all.append(trend_strength)
```

```python
df_all = pd.DataFrame(df_all).T
```

```python
mean = df_all.mean(axis=1)
mean.name = "mean"
std = df_all.std(axis=1)
std.name = "std"
```

```python
from sts.plots.time_series_plots import plot_multi_time_series
```

```python
qqq = Ticker("QQQ")
qqq_df = qqq.history(period=period)
spy = Ticker("SPY")
spy_df = qqq.history(period=period)
```

```python
spy_trend = trend_ewm(spy_df, short, long)
spy_trend.name = "spy"
```

```python
index = 0
```

```python
plot_multi_time_series(
    [
        mean.iloc[index : index + 500],
        std.iloc[index : index + 500],
        spy_df["Close"].iloc[index : index + 500],
    ],
    name_list=["mean", "std", "spy"],
    yaxis_list=["y1", "y2", "y3"],
    yaxis_left=[True, False, False],
)
```
