```python
from sts.data.database import SymbolTable
from sts.data.yahoo.yhdata import YHData
import datetime
```

```python
symbol_table = SymbolTable()
```

```python
symbol_table = symbol_table.add_symbol(YahooSym="BTC-USD", Symbol="BTCUSD")
symbol_table = symbol_table.add_symbol(YahooSym="IWM", Symbol="IWM")
symbol_table
```

```python
sd = datetime.date(2020, 1, 1)
ed = datetime.date(2023, 9, 1)
```

```python
yh_data = YHData(symbol_table=symbol_table)
```

```python
df = yh_data.get_hist_data_for_symbols(["BTCUSD", "IWM"], sd, ed)
```

```python
df = df.dropna()
```

```python
df.apply(lambda x: x / x.iloc[0], axis=0)
```
