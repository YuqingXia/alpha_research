# Introduction

This is the baseline of trending following system purely based on the trend_score to test the predition power

```python
from sts.dio.equity import Ticker
import os
from sts.quant.candle import Candle
from sts.quant.ta.trend import get_trend_score
from sts.stsig.sigscale import scale_and_clip_signal
import pandas as pd
from sts.system.accounts.pnl_calculators.pnl_SR_cost import PnLCalculationWithSRCosts
from sts.system.accounts.curves.account_curve import AccountCurve

pd.options.plotting.backend = "plotly"
```

```python
spy_ticker = Ticker(
    "QQQ",
    data_mode=Ticker.DATAMODE.LOCAL,
    local_data_folder=os.path.expanduser("../../data/market_data"),
)
```

```python
df = Candle(spy_ticker.history(period="10y"))
```

```python
trend_score = get_trend_score(df.log(), short=10)
```

```python
scaled_signal = scale_and_clip_signal(trend_score)
```

```python
avgPosition = pd.Series(1.0, index=scaled_signal.index)
```

```python
pnl_calculation = PnLCalculationWithSRCosts(scaled_signal, df["Close"], SR_cost=0.0, average_position=avgPosition)
```

```python
pnl_calculation.as_pd_series().cumsum().plot()
```
