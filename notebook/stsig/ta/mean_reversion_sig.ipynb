### Intermarket correlation divergence

```python
from sts.dio.equity import TickerDatabase as Ticker
from sts.quant.ta.rv_stats import get_trend_corr_strength
from sts.quant.candle import Candle
import pandas as pd
from sts.plots.time_series_plots import plot_multi_time_series
```

```python
high_correlated_tickers = ["SPY", "QQQ", "IWM", "XLY", "XLI"]
ohlc_dict = {}
for sym in high_correlated_tickers:
    ticker = Ticker(sym)
    df = Candle(ticker.history(period="10y")).log()
    ohlc_dict[sym] = df
```

```python
corr_strength_df = get_trend_corr_strength(ohlc_dict, short=10)
```

```python
lower_percentile = corr_strength_df["mean"] - corr_strength_df["dispersion"]
```

```python
pd.options.plotting.backend = "plotly"
plot_multi_time_series(
    [ohlc_dict["SPY"]["Close"], corr_strength_df["dispersion"]],
    yn_list=["y1", "y2"],
    ypos_left=[True, False],
    ypos_shift=[0, 0.1],
)
```
