```python
from sts.stsig.trending.kqtrend.signal import KQSignal
from sts.data.ticker import Ticker
from sts.quant.candle import Candle
import pandas as pd
```

```python
ticker = Ticker('us/stock.yml')
kqsignal_fn = KQSignal()
```

```python
ticker.top_trading_symbols()
```

```python
df = ticker.history("AAPL", start="2022-01-01", end="2025-05-01")
df.index = pd.to_datetime(df['ts'].dt.tz_convert('America/New_York').dt.date)
signal_dict = kqsignal_fn.get_signal(df)
for t in signal_dict.values():
    print(t[t])
```
