### Data

```python
symbol = 'TSLA'
if False:
    stock = Ticker('us/stock.yml')
    df = stock.history(symbol, '2022-01-01', '2025-12-20')
    df = df.set_index('ts')
else:
    df = pd.read_parquet(Path(f'~/tmp/{symbol}').expanduser())
df2 = Candle(df).log()
df2.plot()
```

### Consolidating
