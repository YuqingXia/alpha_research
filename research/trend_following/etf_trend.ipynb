```python
import pandas as pd
from datetime import datetime, date
```

```python
td = date(2026, 1, 12)
all_df = None
for interval in ['1d', '1W', '1ME']:
    stock_df = pd.read_parquet(f'/home/yuqing42/data/stsdata/research/us_stock/{interval}/{td.strftime('%Y%m%d')}')
    stock_df = stock_df.set_index('symbol')
    if all_df is None:
        all_df = stock_df
    else:
        all_df = pd.merge(all_df, stock_df, left_index = True, right_index = True, how = 'outer')
# etf_df = pd.read_parquet('/home/yuqing42/data/stsdata/research/us_etf/1d/20260112')
```
